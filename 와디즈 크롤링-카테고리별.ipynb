```python
n_news = []
n_comment = []
n_suppoter = []
t_purpose = []
t_term = []
t_percent = []
t_amount = []
n_like = []
n_img = []

driver = webdriver.Chrome("/Users/sgvin/chromedriver.exe")

for path in df1['url']:
    path = 'https://www.wadiz.kr' + path
    driver.get(path)
    time.sleep(0.5)
    soup2=BeautifulSoup(driver.page_source,"html.parser")
    time.sleep(0.5)
    # 새소식 / 커뮤니티 댓글 / 후원자 개수
    n_news.append(soup2.find_all('a', {'class':'tab-link'})[3].get_text()) # 새소식 개수
    n_comment.append(soup2.find_all('a', {'class':'tab-link'})[4].get_text()) # 커뮤니티 댓글 개수
    n_suppoter.append(soup2.find_all('a', {'class':'tab-link'})[5].get_text()) # 후원자 개수

    # 목표 금액 /  펀딩 기간
    t_purpose.append(soup2.find_all('p', {'style':'color:#00b2b2;font-size:13px;line-height:20px;margin-bottom:6px;'})[0].find_all('span')[0].get_text())
    t_term.append(soup2.find_all('p', {'style':'color:#00b2b2;font-size:13px;line-height:20px;margin-bottom:6px;'})[0].find_all('span')[1].get_text())

    # 달성률
    t_percent.append(soup2.find_all('p', {'style':'margin-bottom: 16px;'})[0].find('strong').get_text())
    
    # 달성금액
    t_amount.append(soup2.find('p', {'class':'total-amount'}).get_text())

    # 좋아요 개수
    n_like.append(soup2.find('em', {'class':'cnt-like'}).get_text())

    # 본문의 이미지 개수 
    n_img.append(len(soup2.find_all('img',{'class':'fr-fic fr-dib lazyload'})))
    
    driver.execute_script("window.open();")
    driver.switch_to.window(driver.window_handles[0])
    driver.close()
    driver.switch_to.window(driver.window_handles[-1])
```
